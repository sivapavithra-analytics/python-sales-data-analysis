# Data Visualization

```python
# Revenue by Order

# Total sales column (adjust if column name differs)
if 'SALES' in df.columns:
    plt.figure()
    df['SALES'].hist(bins=20)
    plt.title('Distribution of Sales')
    plt.xlabel('Sales Amount')
    plt.ylabel('Frequency')
    plt.show()
```

# Trend Analysis

```python
print(df['ORDERDATE'].dtype)
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')
```

```python
# Sales trend over time (if ORDERDATE exists)
if 'ORDERDATE' in df.columns and 'SALES' in df.columns:
    sales_over_time = df.groupby(df['ORDERDATE'].dt.to_period('M'))['SALES'].sum()
    sales_over_time.plot()
    plt.title('Monthly Sales Trend')
    plt.xlabel('Month')
    plt.ylabel('Total Sales')
    plt.show()
```
